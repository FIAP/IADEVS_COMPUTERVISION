III. Conversão de Imagem para Grayscale


Conversão de Cores


Converter uma imagem colorida para grayscale é uma operação comum em visão computacional. Isso pode ser feito
facilmente com OpenCV.

```python
img = "/content/FOTO.jpg"
```

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread(img)

# Converter a imagem para grayscale
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Exibir a imagem grayscale
plt.imshow(imagem_gray, cmap='gray')
plt.axis('off')  # Desabilita os eixos
plt.show()
```

IV. Redimensionamento de Imagem


Alterar a Resolução


Redimensionar uma imagem pode ser necessário para padronizar o tamanho das imagens de entrada.

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread(img)

# Redimensionar a imagem
largura = 800
altura = 600
dimensoes = (largura, altura)
imagem_redimensionada = cv2.resize(imagem, dimensoes,
                                   interpolation=cv2.INTER_AREA)

# Converter a imagem de BGR para RGB
imagem_redimensionada_rgb = cv2.cvtColor(imagem_redimensionada,
                                         cv2.COLOR_BGR2RGB)

# Exibir a imagem redimensionada
plt.imshow(imagem_redimensionada_rgb)
plt.axis('off')  # Desabilita os eixos
plt.show()
```

V. Aplicação de Filtros para Suavização de Imagens

Suavização com Filtro Gaussiano

Suavizar uma imagem ajuda a reduzir o ruído e pode melhorar a detecção de bordas.

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread(img)

# Aplicar suavização com filtro Gaussiano
imagem_suavizada = cv2.GaussianBlur(imagem, (15, 15), 0)

# Converter a imagem de BGR para RGB
imagem_suavizada_rgb = cv2.cvtColor(imagem_suavizada, cv2.COLOR_BGR2RGB)

# Exibir a imagem suavizada
plt.imshow(imagem_suavizada_rgb)
plt.axis('off')  # Desabilita os eixos
plt.show()
```

VI. Detecção de Bordas com Canny

Detecção de Bordas

A detecção de bordas é um passo crucial em muitos algoritmos de visão computacional, como detecção de objetos.

```python
import cv2
import matplotlib.pyplot as plt

# Carregar a imagem
imagem = cv2.imread(img)

# Converter a imagem para grayscale
imagem_gray = cv2.cvtColor(imagem, cv2.COLOR_BGR2GRAY)

# Detectar bordas na imagem
bordas = cv2.Canny(imagem_gray, 100, 200)

# Exibir a imagem com bordas detectadas
plt.imshow(bordas, cmap='gray')
plt.axis('off')  # Desabilita os eixos
plt.show()
```

VII. Desenhar Formas na Imagem


Desenho de Retângulo


Podemos desenhar formas como retângulos para destacar áreas de interesse na imagem.

```python
import cv2
import matplotlib.pyplot as plt

imagem = cv2.imread(img)

# Definir as coordenadas e a cor do retângulo
inicio = (300, 5)   # deslocamento do inida esq -> dir, deslocamento de cima -> baixo
fim = (550, 350)   # deslocamento da dir -> esq, deslocamento de baixo -> cima
cor = (255, 0, 0)  # Azul em BGR
espessura = 2

# Desenhar o retângulo na cópia da imagem
imagem_com_retangulo = cv2.rectangle(imagem.copy(),
                                     inicio, fim, cor, espessura)

# Converter a imagem para RGB
imagem_com_retangulo_rgb = cv2.cvtColor(imagem_com_retangulo,
                                        cv2.COLOR_BGR2RGB)

# Exibir a imagem com retângulo
plt.imshow(imagem_com_retangulo_rgb)
plt.axis('off')  # Desabilita os eixos
plt.show()
```

VIII. Salvando a Imagem Processada


Salvar Imagem no Disco


Finalmente, podemos salvar a imagem processada no disco.

```python
import cv2

# Salvar a imagem processada
caminho_salvar = '/content/imagem_processada.jpg'
cv2.imwrite(caminho_salvar, imagem_com_retangulo)
```
